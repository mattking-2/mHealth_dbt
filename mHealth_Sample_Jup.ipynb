```python
import pandas as pd
```

```python
file_path = "/Users/mattking/Documents/Career/Posh/Portfolio/19688730/2-requests-by-session.csv"
df = pd.read_csv(file_path)

print(f"Rows/columns: {df.shape}")
df.head()
```

```python
num_rows = 200000 

# taking random sample of 200k rows 
sample_df = df.sample(n=num_rows, random_state=42)

print(f"Rows/columns: {sample_df.shape}")
sample_df.head()
```

```python
sample_file_path = "/Users/mattking/Documents/Career/Posh/Portfolio/mHealth_req_by_session_sample.csv"
sample_df.to_csv(sample_file_path, index=False)

```
